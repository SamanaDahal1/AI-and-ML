```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

im= Image.open("/content/drive/MyDrive/Lenna.png").convert("L")
a = np.array(im, dtype=float)

plt.imshow(a, cmap='gray')
plt.show()

m = np.mean(a, axis=0)
x = a - m

c = np.cov(x, rowvar=False)

v, e = np.linalg.eig(c)

i = np.argsort(v)[::-1]
v = v[i]
e = e[:, i]

ev = v / np.sum(v)
cv = np.cumsum(ev)

plt.plot(cv)
plt.show()

k = [10, 30, 60, 100]

plt.figure(figsize=(10,8))

for j, n in enumerate(k):
    p = e[:, :n]
    z = np.dot(x, p)
    r = np.dot(z, p.T) + m

    plt.subplot(2,2,j+1)
    plt.imshow(r, cmap='gray')


plt.tight_layout()
plt.show()
```
